```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# 1 Data Preparation
file_path = 'organisations.csv'
df = pd.read_csv(file_path)

df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce')
df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'], errors='coerce')
print(df)

df = df[(df['founded_at'].dt.year >= 1995) & (df['founded_at'].dt.year <= 2015)]

df['market'] = df['market'].str.strip().str.lower().replace('', pd.NA)

df = df.dropna(subset=['market', 'founded_at', 'funding_total_usd', 'country_code', 'status'])

df = df.reset_index(drop=True)

print(df.info())
print(df.head())


# 2 Descriptive Analysis

## 2.1 Number of Startups by Year

startups_by_year = df.groupby(df['founded_at'].dt.year).size()

plt.figure(figsize=(10, 6))
startups_by_year.plot(kind='line', marker='o')
plt.title('Number of Startups Founded by Year (1995-2015)')
plt.xlabel('Year')
plt.ylabel('Number of Startups')
plt.grid(True)
plt.show()

## 2.2 Dominant Markets

### 2.2.1 Top 10 markets by number of startups
top_markets_count = df['market'].value_counts().head(10)

### 2.2.2 Top 10 markets by total funding
top_markets_funding = df.groupby('market')['funding_total_usd'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_markets_count.values, y=top_markets_count.index)
plt.title('Top 10 Markets by Number of Startups (1995-2015)')
plt.xlabel('Number of Startups')
plt.ylabel('Market')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_markets_funding.values, y=top_markets_funding.index)
plt.title('Top 10 Markets by Total Funding (1995-2015)')
plt.xlabel('Total Funding (USD)')
plt.ylabel('Market')
plt.show()

## 2.3 Funding Trends

### 2.3.1 Total funding by year
funding_by_year = df.groupby(df['founded_at'].dt.year)['funding_total_usd'].agg(['sum', 'mean'])

plt.figure(figsize=(10, 6))
funding_by_year['sum'].plot(kind='line', marker='o', label='Total Funding')
plt.title('Total Funding by Year (1995-2015)')
plt.xlabel('Year')
plt.ylabel('Total Funding (USD)')
plt.legend()
plt.grid(True)
plt.show()

## 2.4 No. of Startups vs. Global GDP

file_path = 'World_ Gross domestic product.csv'
gdp = pd.read_csv(file_path)
gdp = gdp.rename(columns={'Variable observation date': 'year'})
gdp = gdp.rename(columns={'Variable observation value': 'gdp value'})
gdp = gdp[(gdp['year'] >= 1995) & (gdp['year'] <= 2015)]
gdp['gdp_trillion'] = gdp['gdp value'] / 1e12
# Ensure 'year' is int64 (in case it's not)
gdp['year'] = gdp['year'].astype('int64')

startup_counts = df.groupby(df['founded_at'].dt.year).size().reset_index(name='startup_count')
startup_counts = startup_counts.rename(columns={'founded_at': 'year'})

startup_gdp = pd.merge(startup_counts, gdp, on='year', how='right')
startup_gdp['startup_count'] = startup_gdp['startup_count'].fillna(0)

plt.figure(figsize=(10,6))
line1,=plt.plot(startup_gdp['year'], startup_gdp['startup_count'], color='tab:blue', marker='o', label='Number of Startups')
plt.xlabel('Year')
plt.ylabel('Number of Startups', color='tab:blue')
plt.tick_params(axis='y', labelcolor='tab:blue')

plt.twinx()
line2,=plt.plot(startup_gdp['year'], startup_gdp['gdp_trillion'], color='tab:red', marker='x', label='Global GDP')
plt.ylabel('Global GDP (Trillion USD)', color='tab:red')
plt.tick_params(axis='y', labelcolor='tab:red')

plt.xticks(ticks=range(1995, 2016, 2))
plt.title('Startups Founded vs Global GDP (1995-2015)')
plt.legend(handles=[line1, line2], loc='upper left')
plt.show()
```
